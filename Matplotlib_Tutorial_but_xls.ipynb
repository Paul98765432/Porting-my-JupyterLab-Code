## Scatter Plot
The scatter plot is a good way to plot non-continuous data. In a scatter plot, the values of two arrays are plotted along two axes as points. You can find the documentation for matplotlib scatter plots [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html). 

You have a lot of control over how to format the points of scatter plots - you can make your points different colors, shapes, opacities, etc, by passing an array for each parameter you want to change. 

```python
xdata = np.linspace(1,10, 10)
ydata = np.random.random(10)

# Plot the "data"
plt.scatter(xdata, ydata, color = 'DarkGreen', marker="*", s = 100)

# Plot another set of 'data'
ydata = np.random.random(10) + 2
plt.scatter(xdata, ydata, color = 'HotPink', marker="^", s = 75, )

# Now change the colors:
ydata = np.random.random(10) + 4
plt.scatter(xdata, ydata, c = ['r','DarkSalmon','Orange','b', 'g', 'c', 'm', 'y', 'k', 'w' ], marker="o", s = 100, ) #I didn't give 10 colors to "c = " - I only gave 3! What's happening?
# Paul Note: I haven't used 10 colors, but no errors show up so it works

# Now randomize the colors:
ydata = np.random.random(10) + 6
cols = np.random.random(10)
cm = matplotlib.colormaps['cool'] # choose a color map
plt.scatter(xdata, ydata, c = cols, cmap = cm, marker="s", s = 100, ) #I didn't give 10 colors to "c = " - I only gave 3! What's happening?


# Add axis labels, and make them large enough to read easily 
plt.xlabel("Trial", fontsize = 15)
plt.ylabel("Random Number", fontsize = 15)

# Make the xticks and yticks large enough to read easily (but do not change their values)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.show()
```

# Line Plot

Line plots show the change of a dependant variable (on the y axis) as mediated by a continuous, independant variable (on the x axis). In Python, they are very easy to create and require only an array for the x data and the y data [(see documentaton here)](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html). 

The functionality of plot is very similar to that of scatter (above), but with less customizability. You can easily change the color of the entire line with the "color=" argument (which is different than scatter's "c="). Also note that if you set "linestyle='None'" as an argument, you will be able to plot only points without the connecting line (you will need to specify a "marker=" type then). 


```python
xdata = np.linspace(0, 2 * np.pi, 100)
ydata = np.sin(xdata)

# Plot the "data"
plt.plot(xdata, ydata, color = 'Navy', linewidth = 2,)

# Add axis labels, and make them large enough to read easily 
plt.xlabel("x", fontsize = 19)
plt.ylabel("sin(x)", fontsize = 19)

# Make the xticks and yticks large enough to read easily (but do not change their values)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.show()
```

# Histograms
Histograms are used to show the distribution of numerical data. They are commonly used to display probability distributions. The y-axis of a histrogram will be the "count / number / fraction / etc" or some measure of the frequency of occurence of an outcome, and the x-axis will be all possible outcomes. You only need one type of data to make a histrogram. 

```python
xdata = np.random.randn(1000) # sample 1000 times from a normal distribution - run help(np.random.randn) if you don't know what this function does


# Plot the "data"
plt.hist(xdata, bins = 20)

# Add axis labels, and make them large enough to read easily 
plt.xlabel("Number chosen", fontsize = 15)
plt.ylabel("Number of times chosen", fontsize = 15)

# Make the xticks and yticks large enough to read easily (but do not change their values)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.show()
```

## Bar Chart
Bar charts ([documentation and options here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)) can be used to present categorical data, or comparisons between and within categories. They are useful to either compare differences between groups, or examine a quantity over time (when measured at distinct time points: ex, Jan., Feb., March, etc.). 

The data for the plot below comes from [this link](https://keplerscience.arc.nasa.gov/k2-approved-programs.html). 

```python
# Define the data
K2_campaign = [15,16,17,18,19]
long_candence_targets = [23279,29888,34398,20419 ,33863]
short_cadence_targets = [118,129,179,235,216]

# Create the bar chart, and stack the two categories on top of each other
plt.bar(K2_campaign, long_candence_targets,label="Long Cadence", color='SkyBlue')
plt.bar(K2_campaign, short_cadence_targets, bottom = long_candence_targets,label="Short Cadence", color='Purple')

# Add axis labels, and make them large enough to read easily 
plt.xlabel("K2 Campaign", fontsize = 15)
plt.ylabel("Number of Targets", fontsize = 15)

# Make the xticks and yticks large enough to read easily (but do not change their values)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

# What if you want to change the xticks to have names (strings) instead?
plt.xticks(K2_campaign, ['C15', 'C16', 'C17', 'C18', 'C19'], fontsize = 14)

# Be careful what you call them...
plt.xticks(K2_campaign, ['C15', 'C16', 'C18', 'C17', 'C19'], fontsize = 14) #oops!


# Add the legend, show the plot
plt.legend(loc=3, fontsize=12) # Where do you want to place the legend on the plot? 1 = upper right, 2 = upper left, etc
plt.show() # display the plot
```

## Pie chart

Pie charts are intended to display data that is part of a whole. The volume of each slice (and the length of the arc) is directly proportional to the value in the marked quantity. Find the [documentation here](https://matplotlib.org/3.1.1/gallery/pie_and_polar_charts/pie_features.html). 

```python
labels = ['Apple', 'Cherry', 'Chocolate', 'Pumpkin', 'I hate pie', 'Lemon']
sizes = [9, 14, 3, 2, 1, 4 ]
plt.title("Survey Results: Best Type of Pie", fontsize = 15)
plt.pie(sizes, labels=labels, autopct='%1.0f%%', shadow=True, startangle=0)
plt.axis('equal')
plt.show()
#Paul Note: there's a weird bug with the pie chart showing what looks like another overlay? I'm not sure where it's coming from.
```

# Plotting previously created data with advanced techniques
Let's try plotting the results of a previously created N-body simulation (which I have called "sample_planet.aei" and saved in this directory), and we will use pandas to load the data and print the first five rows to see what we are working with: 

```python
orbit_data = pd.read_csv("sample_planet.aei", delim_whitespace=True, skiprows=2)
orbit_data[:5]
```

Now that you have read in the data, you can make the plot the same way we did before, but instead of defining your own array as 'xdata', you can take a column from the data.

```python
xdata = orbit_data['Time(years)'].values
ydata = orbit_data['e'].values

# Plot the data
plt.plot(xdata, ydata, color='DarkRed', alpha = 0.7) # alpha makes plot either transparent (=0) or opaque (=1)

# Add axis labels, and make them large enough to read easily 
plt.xlabel("Time (years)", fontsize = 15)
plt.ylabel("Planetary Eccentricity", fontsize = 15)
plt.xlim(0,1e5)

# Make the xticks and yticks large enough to read easily (but do not change their values)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.show()
```

```python
xdata = orbit_data['x(AU)'].values
ydata = orbit_data['y(AU)'].values
zdata = orbit_data['z(AU)'].values

# Create the first subplot
plt.figure(figsize=(7,7))
plt.subplot(2,2,1)
plt.scatter(xdata, ydata)
plt.xticks([],[]) # put nothing on this axis
plt.xlim(-0.5, 0.5)
plt.ylabel("y (AU)", fontsize = 15)
plt.yticks(fontsize = 14)
plt.ylim(-0.12, 0.12)

# Create the second subplot
plt.subplot(2,2,3)
plt.scatter(xdata, zdata)
plt.xlabel("x (AU)", fontsize = 15)
plt.xlim(-0.5, 0.5)
plt.xticks(fontsize = 12)
plt.ylim(-0.5, 0.5)
plt.yticks(fontsize = 14)
plt.ylabel("z (AU)", fontsize = 15)

# Create the last subplot
plt.subplot(2,2,4)
plt.scatter(ydata, zdata)
plt.xticks(fontsize = 12)
plt.ylim(-0.5, 0.5) # match the axis in the plot next to this one
plt.yticks([],[]) # put nothing on this axis
plt.xlim(-0.12, 0.12)
plt.xlabel("y (AU)", fontsize = 15)

# Fix global properties, show plot
plt.subplots_adjust(wspace = 0, hspace=0)
plt.show()
```

Try using the properties of good plots we talked about to fix them... you might need to change the type of plot used, if I didn't use the most effective plot type to visualize the data. Also think about style and ease of understanding!

Visual properties of good plots:
1. Ensure all axes are labeled and have units. 
2. Ensure that all text (including legends, tick markers, etc) is large enough to be readable. 
3. Try to choose colors that will not cause confusion for readers; that is, choose colors different enough that readers can easily discern between them, even if a reader is colorblind. Roughly 5% of the population is red-green colorblind!
4. Keep plots as simple as you can while still getting the main point across. Too much 'noise' on a plot will make it hard for readers to understand your point. 
5. If you put something on the plot, it should have meaning. For example, if you color code your data, then the colors should correspond to something important (not just be random, or not repeat information that is elsewhere on the chart). 
6. Choose a plot style that supports the point you're trying to make about the data.  (Want to show a trend? Line plot or scatter plot. Want to show a ditribution? Histogram. Want to compare values directly? Bar chart, pie chart, line plot...)
    
    
Clarity, Precision, Efficiency:
1. Show the data directly 
2. Present large quantity of data efficiently and in a small space
3. Encourage the eye to compare data
4. Present layers of complexity, but coherent message
5. Serve a clear purpose (caption can help)
6. Connect plots and figures to the text in which they are presented

Integrity:
1. Do not distort data
2. Representation of numbers on a chart / graphic should be directly proportional to true values 
3. Labeling should eliminate ambiguity 
4. Show data variation, not design variation 
5. Normalize data as appropriate 
6. Data should not be used out of context


### Plot 1: Inclination Evolution
The goal of this plot is to show the inclination evolution (inclination is orbital tilt, and the values in the files have units which are listed in the header of the file) of our sample exoplanet integration. There is a single planet, and the inclination evolution can be treated as a continuous variable. 

```python
xdata = orbit_data['Time(years)'].values
ydata = orbit_data['i(deg)'].values

plt.scatter(xdata, ydata, c = ydata) # c sets the color for each point
plt.xlabel("time")
plt.ylabel("i")
plt.legend(["planet"])
plt.show()
```

### Plot 2: Plot orbital periods
