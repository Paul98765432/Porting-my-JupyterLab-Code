```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the Excel file
data_planet = pd.read_excel("ep206103150.xlsx")

# Extract time and flux data
time = data_planet['time']
flux = data_planet['flux']

# Period of the planet (in days)
period = 4.1591

# Calculate phase using the provided period, with the curve aligned at phase 0
phase = ((time - time.min()) / period) % 1

# Adjust phase to center the curve
phase_adjusted = np.where(phase >= 0.5, phase - 1, phase)

# Sort the data based on phase for a cleaner plot
sorted_indices = np.argsort(phase_adjusted)
sorted_phase = phase_adjusted[sorted_indices]
sorted_flux = flux[sorted_indices]

# Phase-folded plot
plt.figure(figsize=(10, 6))
plt.plot(sorted_phase, sorted_flux, marker='o', linestyle='-', color='b')
plt.xlabel('Phase')
plt.ylabel('Flux')
plt.title('Phase-Folded Transit Light Curve for WASP-47b')
plt.grid(True)
plt.show()
```
