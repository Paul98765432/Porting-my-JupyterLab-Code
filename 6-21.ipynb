```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('sample_planet.aei', skiprows = 1, delim_whitespace=True)
```

```python
data
len(data['a(AU)'])
# since sample_planet.aei is more of a dictionary when read in with pandas than it is an array, we use the column name instead of the column number ([column 1] instead of [0])
```

```python
x = data['Time(years)']
y = data['e']
plt.plot(x,y)
plt.xlim(0,10000)
plt.xlabel('Time In Years')
plt.ylabel('Eccentricity of Orbit')
plt.show()
```
